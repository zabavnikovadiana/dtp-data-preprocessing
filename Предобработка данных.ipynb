### <a class="anchor" id="5"></a> <span style="color:#FF00FF"> Преобразование типов данных </span>

```python
df['date_time'] = df['crash_date'].astype(str) + 'T' + df['crash_time'].astype(str) 
# сначала мы объединяем два столбца с разделителем "Т" (требуется в задании), меняя тип данных на строковый (ранее тип данных столбца "crash_date" - целочисленный), так как сложение строк возможно только со строками;
```

```python
df.head() # проверяем добавление нового столбца "date_time" в датасет
```

```python
df['date_time'] = pd.to_datetime(df['date_time'], format='%Y%m%dT%H:%M:%S') # преобразует строковый тип данных в формат даты и времени
```

```python
df.info() # проверяем, что тип данных "date_time" изменился на "datetime64[ns]"
```

```python
df['participants_amount'] =  pd.to_numeric(df['participants_amount'], errors='coerce') 
# преобразует строковый тип данных в числовой, добавляя "errors='coerce'" (т.к. данные смешаные, присутствуют как числа (1, 2, 3 и т.д.), так и "больше 25"), "errors='coerce'" меняет строковый тип данных на "NaN" 
```

```python
df['participants_amount'] = df['participants_amount'].fillna(0) #заполняем 0 пропуски, сделанные принудительно выше
```

```python
df['participants_amount'] = df['participants_amount'].astype('int') # преобразуем 
```

```python
df.info() # осуществляем проверку замены типа дынных столбца 'participants_amount' с "object" на "int64"
```

#### <span style="color:deepskyblue">Вывод</span>
1. Поработав над типами данных, мы преобразовали столбец "date_time" в datetime64[ns] (из строкового типа).
2. "participants_amount" преобразован в int64 (из object).
3. Теперь датасет состоит из след. типов данных:
   строковый формат: 8 столбцов;
   дата/время: 1 столбец;
   целочисленный тип данных: 6 столбцов.

### <a class="anchor" id="7"></a> <span style="color:#FF00FF"> Добавление столбцов </span>
